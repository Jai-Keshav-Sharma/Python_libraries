```python
import numpy as np
import pandas as pd
```

### Pandas has two types of data structures:
 Series and Dataframe
 1. Series: it is a 1-D array with indexes, it stores a single row or column of data in a DataFrame
 2. DataFrame: it is a spreadsheet like 2-D structure representing rows, each of which contains one or multiple columns 

```python
dict1 = {"name":['harry', 'rohan', 'skillf', 'shubh'],
         "marks":[92 , 34, 24, 17],
         "cities": ['rampur', 'kolkata', 'bareilly', 'anatrctica']
         }
```

### creating dataframe

```python
df = pd.DataFrame(dict1)
df
```

```python
# to show first few elememts of dataframe
df.head(2)
```

```python
# to show last few elememts of dataframe
df.tail(2)
```

```python
# to statistically analyse numerical data present in dataFrame
df.describe()
```

#### copying dataframe

```python
newdf2 = newdf
newdf2[0][0] = 999
newdf
```

```python
newdf2 = newdf.copy()
newdf2[0][0] = 1000
newdf
```

**the warning shown above is because when we update a value in dataFrame like this, it is not always guaranteed that a copy of dataframe is made.
In case a view of dataframe is made, the dataframe is not copied, and the original dataframe is modified.
else in case of copy, the original dataframe remains intact.**


### how to change data in DataFrame ?

```python
## use of loc[row_ind, col_ind]
newdf.loc[0, 0] = 99
```

```python
newdf.head()
```

#### deleting a column

```python
newdf.drop(4, axis=1)
```

#### more with loc

```python
newdf.loc[[1, 2], [2, 3]]
```

```python
newdf.loc[:, [2, 3]]
```

```python
newdf.loc[[1, 2], :]
```

```python
newdf.loc[newdf[0] < 10]
```
